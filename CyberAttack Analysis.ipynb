# Analysis of Cyber Attacks

## Types Of Attacks Covered
1. Reconnaissance - attacker gathers info about computer network to evade security controls
2. Fuzzers - attacker attempts to uncover security loopholes by feeding massive random data entry 
3. Analysis - penetrates web applications through ports, emails, web scripts
4. Backdoor - technique to avoid normal authentication to ensure unauthorized remote access to a device
5. Exploit - exploits vulnerability caused by involuntary or unsuspected behavior on host or network 
6. Generic - attempts against block encryption using hash functions for collision 
7. Shellcode - attacker penetrates a small piece of code from shell to control compromised machine 
8. Worm - attack replicates malicious script to spread it to other computers
9. DoS (Denial of Service) - an intrusion that disrupts computer resources to be extremelyy busy to prevent unauthorized requests from accessing a device

### Performing T test

```python
for attack in list(newdf['Attack category'].unique()):
    df_attack = newdf[newdf['Attack category'] == attack].copy()
    statistic, pvalue = stats.ttest_ind(df_attack['Source Port'], df_attack['Destination Port'], equal_var=False)
    print('p-value in T-test for ' + attack + ' attack: ' + str(pvalue))
```

The -values of all but one attack category are very close to 0.0. This means that the attacks have been directed to the specific ports, except for the Shellcode attacks, whose null hypothesis cannot be rejected. For this type of attack there is a defined randomness, which means that the source and destination ports have similar averages.

```python
df_crosstab = pd.crosstab(newdf['Attack category'], newdf['Destination Port'])
df_crosstab
```

### Performing Chi-square Test
Testing whether the variations we observe reflect real differences or are just a by product of randomness

```python
chi2, p_value, dof, expected = chi2_contingency(df_crosstab)
print("p-value of Chi-square test for Attack category vs. Destination Port =", p_value)
```
